### 2. Research each of these features and explain the relationships between them (mfcc, Zero Crossing Rate  , mel-spectrogram and chroma features) .

MFCC (Mel Frequency Cepstral Coefficients): These are coefficients that collectively represent the short-term power spectrum of a sound. MFCCs are derived from the Fourier Transform and are often used in speech and audio processing tasks due to their effectiveness in capturing relevant features of the sound signal.

Zero Crossing Rate: This feature measures the rate at which the signal changes its sign, which can provide information about the frequency content and periodicity of the signal.

Mel-Spectrogram: A spectrogram is a visual representation of the spectrum of frequencies of a signal as it varies with time. Mel spectrograms are spectrograms with frequencies converted to the mel scale, which approximates the human auditory system's response to frequencies.

Chroma Features: These features represent the energy distribution of pitch classes in a sound signal, which can be useful for tasks involving music and harmony analysis.

```python
data_dir = "/home/amirali/Desktop/AI/CA2/recordings"
digit_labels = list(range(10))

def extract_mfcc(audio_file, n_mfcc=13):
    (rate, sig) = wav.read(audio_file)
    mfcc_feat = mfcc(sig, rate, numcep=n_mfcc)
    return mfcc_feat
```

```python
def extract_mfcc_for_digit(digit, data_dir, n_mfcc=13):
    digit_files = [file for file in os.listdir(data_dir) if file.startswith(f"{digit}_")]
    mfccs_per_digit = []
    for file in digit_files:
        file_path = os.path.join(data_dir, file)
        y, sr = librosa.load(file_path)
        mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
        mfccs_per_digit.append(mfccs)
    return np.concatenate(mfccs_per_digit, axis=1)  # concatenate mfcc arrays along axis 1

def plot_mfcc_heatmap_for_digit(mfccs, digit):
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(mfccs, x_axis='time')
    plt.colorbar()
    plt.title(f"MFCCs for digit {digit}")
    plt.tight_layout()
    plt.show()

for digit in range(10):
    mfccs = extract_mfcc_for_digit(digit, data_dir)
    plot_mfcc_heatmap_for_digit(mfccs, digit)
```

## Introduction to HMM :

### 1. Explain what is meant by States and Observation? What are the states in this exercise and how is observation obtained?
States: It means our hidden states, which our goal is to predict in which hidden state we are based of the observations we saw until now. Each state has a chance of moving to another state or staying there at each step. in this exercise states are the digits 0 through 9 that are being spoken by different speakers. Each state corresponds to a specific digit.

Observation: It means our evidences, which are our noisy observatin from the world and we only get to see those in our HMM.Difference between Markov and Hidden Markov is in observations.The observations are the MFCC coefficients extracted from the audio recordings of each spoken digit. These MFCC coefficients capture the acoustic features of the speech signals, which can be used to infer the underlying states (the spoken digits) through an HMM-based speech recognition system.

### 2. HMM models can be classified based on the degree of dependence between hidden states, the model you will implement in this exercise is an First-Order HMM model. Check out the reason for its name as well as its features and explain its difference from other models in this category.

First Order HMM also called the first-order Markov model is in a shape that each hidden state has one observation connected to it and hidden states are connected in the form of a simple Markov chain so that's why it is called first-order. This model is simple and only considers that the current state only depends on the past. It has a low computational cost. we can model many types of situations and works well when our sequence has a short-term dependency with this type of modeling. More complex models can be utilized to handle situations where our relations are not really simple. so in summary we can see Memoryless Transitions, Simplicity, and Stationarity in this modeling system.

There are also other types of modeling like :
Second-Order HMM: The probability of the current state would depend on the two immediately preceding states
Third-Order HMM: The probability of the current state would depend on the three immediately preceding states

### 3. Research about HMM and explain what phenomena this model is suitable for investigating and analyzing? Explain why this is the case.
HMM  models  are  well-suited for analyzing and modeling phenomena characterized by sequential data and underlying hidden states. They are useful when working with sequential data like a series of events or a time series. 

some example of using HMM:

- Speech Recognition : As speech can be modeled as a time series of sound waves like how we can see in this project.

- Natural Language Processing: Model the sequential nature of language and capture the dependencies between different words.

- Weather Forecasting: Can model weather sequences where states represent specific weather conditions and the observations might be related atmospheric measurements like temperature or wind.

### 4. The HMM model, like any other model, has advantages and disadvantages that make it special. Examine the advantages and disadvantages of this model and briefly explain each one.
Advantages:

- Flexibility of hmm has made it popular for wide range of problems and it is adapted to different domains and types of input data.
- has a high interpretability and we can easily understand it.
- Low computational cost.

Disadvatages:

### 5. There are different types of HMM models, research them and briefly review a few.
- Continuous:The observations are continuous distributions, often Gaussian. Continuous HMMs are used in applications where the data are real-valued, such as in econometrics or speech processing for modeling spectral coefficients.
- Mixture:It is a mixture of Gaussian distributions to model the observation probabilities. They are particularly useful in speech and audio processing where a single Gaussian distribution is not sufficient to model the complex variations in feature vectors.
- Hierarchical: Used when there exist a hierarchy in the sequence and each state captures some level of it.

```python
def hmm_train(file_paths):
    hmm_model_results = []
    test_list = []
    
    for digit in digit_labels:
        x_train = np.array([])
        x_train_paths, x_test_paths = train_test_split(file_paths[digit], test_size=0.3, random_state=42)
        
        for k in range(len(x_test_paths)):
            test_list.append((x_test_paths[k], digit))

        for audio_file in x_train_paths:
            mfcc_features = extract_mfcc(audio_file)
            if x_train.size == 0:
                x_train = mfcc_features
            else:
                x_train = np.append(x_train, mfcc_features, axis=0)
        
        model = hmm.GaussianHMM(n_components=4, covariance_type='diag', n_iter=12)
        model.fit(x_train)
        hmm_model_results.append(model)

    return hmm_model_results, test_list


def evaluate(hmm_model_results, test_list, mode):
    accu = 0
    results = []

    for path, label in test_list:
        mfcc_features = extract_mfcc(path)
        model_scores = [model.score(mfcc_features) for model in hmm_model_results]
        predicted_digit = np.argmax(model_scores)
        results.append((predicted_digit, label))
        
        if label == predicted_digit:
            accu += 1 

    c_matrix = confusion_matrix([x[1] for x in results], [x[0] for x in results])

    plt.figure(figsize=(8, 6))
    sns.heatmap(c_matrix, annot=True, cmap='Reds', fmt='d', xticklabels=digit_labels, yticklabels=digit_labels)
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix target var = digital label')

    accu /= len(test_list)
    print(f"Accuracy: {accu}")

# file paths for each spoken digit
file_paths = {digit: [] for digit in digit_labels}
for file in os.listdir(data_dir):
    if file.endswith(".wav"):
        digit_label = int(file.split("_")[0])  #for extracting digit label from file name
        file_paths[digit_label].append(os.path.join(data_dir, file))

# Train 
hmm_model_results, test_list = hmm_train(file_paths)

evaluate(hmm_model_results, test_list, mode="library hmm")
```

```python
def train_models_by_speaker(data_dir):
    speakers = set()
    speaker_files = {}
    for file in os.listdir(data_dir):
        if file.endswith(".wav"):
            speaker_name = file.split("_")[1]
            speakers.add(speaker_name)
            if speaker_name not in speaker_files:
                speaker_files[speaker_name] = []
            speaker_files[speaker_name].append(os.path.join(data_dir, file))
    
    hmm_models = {}
    for speaker in speakers:
        mfcc_features = np.array([])
        for file_path in speaker_files[speaker]:
            features = extract_mfcc(file_path)
            if mfcc_features.size == 0:
                mfcc_features = features
            else:
                mfcc_features = np.concatenate((mfcc_features, features), axis=0)
        model = hmm.GaussianHMM(n_components=4, covariance_type="diag", n_iter=100)
        model.fit(mfcc_features)
        hmm_models[speaker] = model
    
    return hmm_models, speaker_files

def evaluate_models(models, speaker_files):
    confusion_matrix = np.zeros((len(models), len(models)), dtype=int)
    total_samples = 0
    correct_predictions = 0
    for speaker, files in speaker_files.items():
        for file_path in files:
            total_samples += 1
            mfcc_features = extract_mfcc(file_path)
            scores = {target_speaker: model.score(mfcc_features) for target_speaker, model in models.items()}
            predicted_speaker = max(scores, key=scores.get)
            confusion_matrix[list(models.keys()).index(predicted_speaker)][list(models.keys()).index(speaker)] += 1
            if predicted_speaker == speaker:
                correct_predictions += 1

    accuracy = correct_predictions / total_samples

    precision = np.zeros(len(models))
    for i in range(len(models)):
        if confusion_matrix[i][i] > 0:
            precision[i] = confusion_matrix[i][i] / sum(confusion_matrix[i])

    print("Confusion Matrix:")
    print(confusion_matrix)
    print(f"Accuracy: {accuracy}")
    print("Precision:")
    for i, speaker in enumerate(models.keys()):
        print(f"Speaker {speaker}: {precision[i]}")

    plt.imshow(confusion_matrix, cmap='GnBu')
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title("Confusion Matrix")
    plt.colorbar()
    plt.show()

# Train 
models, speaker_files = train_models_by_speaker(data_dir)
evaluate_models(models, speaker_files)
```

```python
class HMM:
    def __init__(self, num_hidden_states):
        self.num_hidden_states = num_hidden_states
        self.rand_state = np.random.RandomState(1)

        self.initial_prob = self._normalize(self.rand_state.rand(1, self.num_hidden_states))
        self.transition_matrix = self._stochasticize(self.rand_state.rand(self.num_hidden_states, self.num_hidden_states))

        self.mean = None
        self.covariances = None
        self.num_dimensions = None

    def _forward(self, observation_matrix):
        log_likelihood = 0.
        T = observation_matrix.shape[1]
        alpha = np.zeros(observation_matrix.shape)
        for t in range(T):
            if t == 0:
                alpha[:, t] = self.initial_prob[:] * observation_matrix[:, t]
            else:
                alpha[:, t] = np.dot(alpha[:, t - 1], self.transition_matrix) * observation_matrix[:, t]

            alpha_sum = np.sum(alpha[:, t])
            alpha[:, t] /= alpha_sum
            log_likelihood += np.log(alpha_sum)

        return log_likelihood, alpha

    def _backward(self, observation_matrix):
        T = observation_matrix.shape[1]
        beta = np.zeros(observation_matrix.shape)

        beta[:, -1] = np.ones(observation_matrix.shape[0])

        for t in range(T - 1)[::-1]:
            beta[:, t] = np.dot(observation_matrix[:, t + 1], self.transition_matrix) * beta[:, t + 1]
            beta[:, t] /= np.sum(beta[:, t])

        return beta

    def _state_likelihood(self, obs):
        obs = np.atleast_2d(obs)
        B = np.zeros((self.num_hidden_states, obs.shape[1]))

        for s in range(self.num_hidden_states):
            np.random.seed(self.rand_state.randint(1))
            mean = self.mean[:, s]
            cov = self.covariances[:, :, s]
            B[s, :] = multivariate_normal.pdf(obs.T, mean=mean, cov=cov, allow_singular=True)

        return B

    def _normalize(self, x):
        return (x + (x == 0)) / np.sum(x)

    def _stochasticize(self, x):
        return (x + (x == 0)) / np.sum(x, axis=1)

    def _em_init(self, obs):
        if self.num_dimensions is None:
            self.num_dimensions = obs.shape[0]
        if self.mean is None:
            subset = self.rand_state.choice(np.arange(self.num_dimensions), size=self.num_hidden_states, replace=False)
            self.mean = obs[:, subset]
        if self.covariances is None:
            self.covariances = np.zeros((self.num_dimensions, self.num_dimensions, self.num_hidden_states))
            self.covariances += np.diag(np.diag(np.cov(obs)))[:, :, None]

        return self

    def _em_step(self, obs):
        obs = np.atleast_2d(obs)
        T = obs.shape[1]

        B = self._state_likelihood(obs)

        log_likelihood, alpha = self. _forward(B)

        beta = self._backward(B)

        xi_sum = np.zeros((self.num_hidden_states, self.num_hidden_states))
        gamma = np.zeros((self.num_hidden_states, T))

        for t in range(T - 1):
            partial_sum = (alpha[:, t] * self.transition_matrix.T).T * B[:, t + 1] * beta[:, t + 1]           
            xi_sum += self._normalize(partial_sum)

            partial_g = alpha[:, t] * beta[:, t]
            gamma[:, t] = self._normalize(partial_g)

        partial_g = alpha[:, -1] * beta[:, -1]
        gamma[:, -1] = self._normalize(partial_g)


        expected_prior = gamma[:, 0]

        gamma_sum = np.repeat(np.sum(xi_sum, axis=1)[:, np.newaxis], axis=1, repeats=self.num_hidden_states)
        expected_transition = self._stochasticize(xi_sum / gamma_sum)

        expected_covariances = np.zeros((self.num_dimensions, self.num_dimensions, self.num_hidden_states))
        expected_covariances += .01 * np.eye(self.num_dimensions)[:, :, None]

        gamma_state_sum = np.sum(gamma, axis=1)
        gamma_state_sum = gamma_state_sum + (gamma_state_sum == 0)

        expected_mean = np.zeros((self.num_dimensions, self.num_hidden_states))
        for s in range(self.num_hidden_states):
            gamma_obs = obs * gamma[s, :]
            expected_mean[:, s] = np.sum(gamma_obs, axis=1) / gamma_state_sum[s]

        self.initial_prob = expected_prior
        self.mean = expected_mean
        self.covariances = expected_covariances
        self.transition_matrix = expected_transition

        return log_likelihood

    def train(self, obs, num_iterations=1):
        for i in range(num_iterations):
            self._em_init(obs)
            self._em_step(obs)
        return self

    def score(self, obs):
        B = self._state_likelihood(obs)
        log_likelihood, _ = self._forward(B)
        return log_likelihood
```

```python
def scratch_hmm(file_paths):
    hmm_model_results = []
    test_list = []
    for digit in digit_labels:
        x_train = np.array([])
        x_train_paths, x_test_paths = train_test_split(file_paths[digit], test_size=0.3, random_state=42)
        
        for k in range(len(x_test_paths)):
            test_list.append((x_test_paths[k], digit))

        for audio_file in x_train_paths:
            mfcc_features = extract_mfcc(audio_file)
            if x_train.size == 0:
                x_train = mfcc_features
            else:
                x_train = np.append(x_train, mfcc_features, axis=0)
        model = HMM(4)
        model.train(x_train.T / np.amax(x_train.T), 2)
        hmm_model_results.append(model)
    return hmm_model_results, test_list

def evaluate(hmm_model_results, test_files):
    accu = 0
    results = []
    confusion_matrix = np.zeros((10, 10), dtype=int)
    digit_counts = np.zeros(10, dtype=int)

    for path, label in test_files:
        mfcc_features = extract_mfcc(path)
        model_scores = [model.score(mfcc_features.T / np.amax(mfcc_features.T)) for model in hmm_model_results]
        predicted_digit = np.argmax(model_scores)
        results.append((predicted_digit, label))
        confusion_matrix[predicted_digit][label] += 1
        
        if label == predicted_digit:
            accu += 1 
        
        digit_counts[label] += 1

    accuracy = accu / len(test_files)

    precision = np.zeros(10)
    for i in range(10):
        if confusion_matrix[i][i] > 0 and digit_counts[i] > 0:
            precision[i] = confusion_matrix[i][i] / digit_counts[i]
        else:
            precision[i] = 0  # precision to 0 or N/A if denominator is 0

    print("Confusion Matrix (target bar = digital Label):")
    print(confusion_matrix)
    print(f"Accuracy: {accuracy}")
    print("Precision:")
    for i in range(10):
        print(f"Digit {i}: {precision[i]}")

    plt.imshow(confusion_matrix, cmap='Greys')
    plt.xlabel("Actual")
    plt.ylabel("Predicted")
    plt.title("Confusion Matrix (target bar = digital Label)")
    plt.xticks(ticks=np.arange(10), labels=digit_labels)
    plt.yticks(ticks=np.arange(10), labels=digit_labels)
    plt.colorbar()
    plt.show()

    #plt.bar(np.arange(10), digit_counts, tick_label=digit_labels)
    #plt.xlabel("Digit")
    #plt.ylabel("Count")
    #plt.title("Data Distribution")
    #plt.show()

#training and testing sets
test_list = []
for digit, paths in file_paths.items():
    _, test_files = train_test_split(paths, test_size=0.3, random_state=42)
    test_list.extend([(file, digit) for file in test_files])

hmm_model_results, _ = scratch_hmm(file_paths) 
evaluate(hmm_model_results, test_list)

```

### 1&3.There are many criteria to measure and evaluate the performance of models; For example: F1, Recall, Precision, Accuracy. Score  Research each of the above criteria and explain how to calculate each one. - Explain how each of the criteria evaluates the model.


- Accuracy: It measures the overall correctness of the model predictions. It is calculated as the ratio of the number of correct predictions to the total number of data. It can be misleading when there is an imbalanced dataset:

$$\ \text{Accuracy} = \frac{\text{Number of Correct Predictions}}{\text{Total Number of Predictions}} =\frac{\text{True Positive} + \text{True Negative}}{\text{True Positive} + \text{True Negative} + \text{False Positive} + \text{False Negative}} \$$

- Precision : Quantifies the percentage of positive identifications by the model that are indeed accurate and valid. It is determined by the fraction of true positives relative to the aggregate of true positives and false positives.particularly in the context of binary classification problems.

$$ \text{Precision} = \frac{\text{True Positives}}{\text{True Positives + False Positives}} \$$

- Recall:Assesses the ability of a model to correctly capture all actual instances of a particular class; it‚Äôs the fraction of true positives out of the total actual positives, measuring how often the model misses true cases. In the context of HMMs, it reflects the model‚Äôs accuracy in identifying the correct sequence of states.it should be considered alongside other metrics (such as precision and F1 score) to comprehensively evaluate a model's performance in different aspects.

$$ \text{Precision} = \frac{\text{True Positives}}{\text{True Positives + False Negatives}} \$$

- F1-score:Harmonizes precision and recall into a single metric by taking their harmonic mean; it gives a balance between a model‚Äôs ability to accurately label true positives without wrongly labeling negatives.ombines precision and recall into a single value, providing a balanced measure of a model's performance in binary classification problems. It is especially useful when seeking an equilibrium between the precision and recall performance of a classifier.

$$ \text{F1 score} = \frac{2 \times (\text{Precision} \times \text{Recall})}{\text{Precision} + \text{Recall}} $$


### 2.Is it a challenge to calculate the mentioned criteria for this project which is class-multi? If yes, what solution do you have for it?

yes its challengive ;Macro, micro, and weighted averaging are commonly used techniques to calculate evaluation criteria for multi-class classification tasks like speaker recognition. Here's how they can be applied:

Macro Averaging: This method calculates the metric independently for each class and then takes the average across all classes. It treats all classes equally, regardless of their frequency in the dataset.

Micro Averaging: Micro averaging aggregates the contributions of all classes before computing the metric. It considers each instance equally, making it suitable for imbalanced datasets where some classes have significantly fewer samples.

Weighted Averaging: Weighted averaging assigns weights to each class based on their frequency in the dataset. It computes the metric for each class and then takes a weighted average, with weights proportional to the class frequency. This approach ensures that classes with more samples have a greater impact on the overall metric.


### 4. State the difference between Recall and Precision and explain why each alone is not sufficient to evaluate the model? For each, give an example where this criterion has a high value but the model does not perform well.

Precision and Recall are vital metrics for evaluating classification models. Precision measures the proportion of correctly predicted positive instances out of all instances predicted as positive, while Recall measures the proportion of correctly predicted positive instances out of all truly positive instances.

For example, imagine a dataset with 10 samples, where 7 are positive. If our model correctly predicts all 7 positive samples, Precision would be 0.7, indicating 70% of positive predictions were accurate. Meanwhile, Recall would be 1, as all true positive samples were identified. This highlights the significance of both metrics in evaluating model performance.

Conversely, if our model correctly predicts only one positive sample out of 7, with the rest labeled as negative, Precision would be 1 (no false positives), but Recall would be 0.14 (only 14% of true positives were identified).

Considering both Precision and Recall together provides a comprehensive view of model effectiveness. High Recall but low Precision, as seen in disease screening labeling all tests as "disease present," catches all cases but flags healthy individuals incorrectly. Conversely, high Precision but low Recall, like in identifying specific spam emails, may miss many other spam emails, resulting in poor performance.
